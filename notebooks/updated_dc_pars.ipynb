```python
with open(os.path.join(wd, 'fix.100'), "r") as f:
    updated = []
    for line in f.readlines():
        parnam = line.split()[1].strip().replace("'", "").replace(",", "_")
        if (parnam in df.loc[:, "name"].tolist()):
            new_val = df.loc[df['name'] == parnam, "max"].tolist()[0]
            new_line = f"{new_val:<14.4f}{line.split()[1].strip()}\n"
            updated.append(new_line)
        else:
            updated.append(line)
            
with open(os.path.join(wd, 'fix.100'), "w") as wf:
    wf.writelines(updated)
```
